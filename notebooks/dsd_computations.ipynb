# Notebook for computing all variants of the DSD embeddings

## Generating a sample network

```python
%load_ext autoreload
%autoreload 2
import numpy as np

# Install the networkx package for handy visualization of the network
#!pip install networkx

import networkx as nx
from scipy.spatial.distance import pdist, squareform

A  = np.random.rand(100, 100) 
A /= np.max(A)
A = A + A.T
A = np.where(A < 0.6, 0, A)
```

## Visualize this network

```python
import matplotlib.pyplot as plt

G = nx.from_numpy_matrix(A)
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos = pos)
plt.show()
```

## Compute DSD embedding

```python

from glidetools.algorithm import compute_dsd_embedding
from scipy.spatial.distance import squareform, pdist

X  = compute_dsd_embedding(A)
Dx = squareform(pdist(X)) 
```

## Compute Reduced DSD embedding

```python
from dsd import compute_dsd_reduced_embedding

X_50  = compute_dsd_reduced_embedding(A, dims= 50)
D_x50 = squareform(pdist(X_50)) 
```
